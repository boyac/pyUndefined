```python
import pandas
import time
from datetime import datetime

dfs = pandas.read_html('http://rate.bot.com.tw/xrt?Lang=zh-TW')
```

```python
dfs[0]
```

```python
currency = dfs[0]
```

```python
currency = currency.ix[:,0:5] #選取需要的[row,column]
```

```python
currency.columns = [u'幣別', u'現金匯率-本行買入', u'現金匯率-本行賣出', u'即期匯率-本行買入', u'即期匯率-本行賣出'] #修改 header
```

```python
currency[u'幣別'] = currency[u'幣別'].str.extract('\((\w+)\)')
```

```python
currency.to_excel('yen_rate.xlsx')
#count = 1

#while True:
#    filename = 'yen_rate_' + str(count) + '.xlsx'
#    currency.to_excel(filename)
#    time.sleep(1*60)
#    count += 1
```

```python
import sqlite3

#寫入 yen_rate 資料庫
with sqlite3.connect('currency.sqlite') as db:
    currency.to_sql('yen_rate', con = db, if_exists='append')
```

```python
#讀取 yen_rate 資料庫
with sqlite3.connect('currency.sqlite') as db:
    df = pandas.read_sql_query('select * from yen_rate', con = db)
```
